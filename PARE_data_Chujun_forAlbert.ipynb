```python
def track_df(track_id, frame_rate):

  # get the specific track
  track = data[track_id]
  
  # get the number of frames in the track
  n_frame = track['frame_ids'].shape[0]
  
  # get the first column of the data frame we want: frame ids
  # and convert it from 1d numpy array to 2d numpy array
  frameID = np.reshape(track['frame_ids'], (n_frame,1))
  
  # get the second column of the data frame we want: time
  timeSec = frameID * frame_rate

  # get the bounding boxes, note that there are four columns cx, cy, w, h
  boxes = track['bboxes']

  # concatenate these three columns first
  # concatenate only works for two arrays at a time
  df_output = np.concatenate((frameID, timeSec), axis=1)
  df_output = np.concatenate((df_output, boxes), axis = 1)

  # get the first 24 joints, note that each joint has three columns x, y, z
  # note that I need an initial numpy array of n_frame x 1 to start the for loop
  df_joint = np.reshape(track['joints3d'][:,0,0], (n_frame,1))
  # loop over the 24 joints
  for j in range(24): 
    # loop over the 3 coordinate within each joint
    for c in range(3): 
      temp = np.reshape(track['joints3d'][:,j,c], (n_frame,1))
      df_joint = np.concatenate((df_joint, temp), axis=1)

  # in the end, remember to discard the first column
  df_output_joint = np.delete(df_joint,[0],1)

  # finally, concatenate the df_output with this joint numpy array
  df_output = np.concatenate((df_output, df_output_joint), axis=1)

  return df_output
```

```python
# Define the function that assembles all the tracks of the same target
# this function takes in two arguments track_id_list which is a list of the track ids
# and the argument of frame_duraction which is the length of each frame
def target_df(track_id_list, frame_duration):
  # first, get the data frame of the first track in the track_id_list:
  track_data = track_df(track_id = track_id_list[0], frame_rate = frame_duration)
  # then, loop over and concatenate the other tracks in the track_id_list:
  for id in track_id_list[1:] : 
    track_temp = track_df(track_id = id, frame_rate = frame_duration)
    track_data = np.concatenate((track_data, track_temp), axis=0)
  return track_data
```
